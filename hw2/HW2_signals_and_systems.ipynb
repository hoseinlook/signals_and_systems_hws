```python
def plot_ct_signal(t, x_t,color='b',label='func'):
    plt.plot(t , x_t,color=color,label=label )
    plt.xlabel('t')
    plt.ylabel('x(t)')
```

```python
def plot_dt_signal(n, x_n ,color='r',label='func'):
    
    markerline1, stemlines, _ = plt.stem(n , x_n ,'-.',label=label)
    plt.setp(markerline1, 'markerfacecolor', color)
    plt.xlabel('n')
    plt.ylabel('x[n]')
```

```python
def u(t) : return np.around((np.sign(t)+1)/2+0.00000001)
```

```python
def delta_dt(n):return u(n)-u(n-1)
```

```python
def convolution_dt(t, x ,h):
    K = np.arange(-(10**3),2*10**3, step=1)
    x_t = x(K)
    h_t=h(K)
    
    y_t = sum([x(k)*h(t-k)  for k in K])
    return y_t


def convolution_ct(t, x ,h):
    K = np.arange(-(10**2),10**2, step=0.1)
    x_t = x(K)
    h_t=h(K)
    
    y_t = sum([x(k)*h(t-k)  for k in K])
    return y_t
    
```
